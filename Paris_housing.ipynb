```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

Split Training or testing Data

```python
int_data = df.iloc[:,:-1]
int_data
```

```python
out_data = df["price"]
out_data
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(int_data, out_data, train_size=0.75, random_state=42)
```

```python
print(x_train.shape,x_test.shape) 
print(y_train.shape, y_test.shape)
```

```python
x_train
```

```python
y_test
```

Import Model

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
```

Training Model

```python
linear_model.fit(x_train, y_train)
```

```python
linear_model.intercept_
```

```python
linear_model.coef_
```

```python
y_pred = linear_model.predict(x_test)
y_pred
```

```python
plt.figure(figsize=(15,8))
xp = range(50)
plt.plot(xp,y_pred[0:50],label = 'Predicted')
plt.plot(xp,y_test[0:50],label = 'Actual')
plt.legend()
plt.show
```

Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
rsme = np.sqrt(mse)
rsme
```

```python
r2score = r2_score(y_test, y_pred)
r2score
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print("Train R:", r2_score(y_train, linear_model.predict(x_train)))
print("Test R:", r2_score(y_test, linear_model.predict(x_test)))
```

```python
s_mae = mean_absolute_error(y_test, y_pred)
print('s_mae', s_mae)
s_mse = mean_squared_error(y_test, y_pred)
print('s_mse', s_mse)
s_rmse = np.sqrt(s_mse)
print('s_rmse',s_rmse)
print('R2_score', r2_score(y_test, y_pred))
```
